Imagine you arrive as a tourist in Berlin - via Tegel Airport, Schoenefeld Airport, or Hauptbahnhof. What locations are you most probably going to visit (according to Foursquare data)?

To get started, we first need a access to the Foursquare API. If you don't have access tokens yet, you can register an app [here](https://developer.foursquare.com/) and use the Client ID and the Client Secret to plug them in below. I load them from a Pickle file in memory.

```python
import time
today = time.strftime('%x')
print 'Exploring Berlin\'s Foursquare data on {}.'.format(today)
```

```python
from cPickle import load
import foursquare
import pandas as pd

keys = load(file('foursquare_keys.pkl'))
client = foursquare.Foursquare(client_id=keys['client_id'], client_secret=keys['client_secret'])
```

Next, we define the location bounding box of Berlin. A bounding box is an area defined by two longitudes and two latitudes. All data that we are going to mine from Foursquare is within these geographical bounds.

```python
berlin_bbox = [13.088400, 52.338120, 13.761340, 52.675499]
```

Now, we set the Foursquare venue IDs from which we want to start our crawl. These are the IDs for Tegel Airport, Schoenefeld Airport, and Hauptbahnhof.

```python
# Venue IDs for Tegel, Schoenefeld and Hbf
to_crawl = ['4adcda92f964a520f14b21e3', '4adcda91f964a520ae4b21e3', '4a1c8506f964a520457b1fe3']
```

The way we are going to crawl the Foursquare data is like this:

- We start retrieving data from Tegel, Schoenefeld, and Hauptbahnhof
- From there, we go to the 5 venues that people have mostly visited after the current one (https://developer.foursquare.com/docs/venues/nextvenues)
- From these 5 venues we go to the 5 next venues
- We keep doing this recursively until a certain depth has been reached or there are no next locations to go to

```python
new_crawl = [] # List of new locations to be crawled
done = [] # List of crawled locations
links = [] # List of tuples that represent links between locations
venues = pd.DataFrame() # Data of crawled locations
```

```python
depth = 10
```

```python
for i in xrange(depth):
    new_crawl = []
    print "Step {}: {} locations and {} links. {} venues to go.".format(i+1, len(venues), len(links), len(to_crawl))
    for v in to_crawl:
        # If venue has not been visited
        if v not in venues:
            res = client.venues(v)
            venues = venues.append(
                pd.DataFrame({"name": res["venue"]["name"], "users": res["venue"]["stats"]["usersCount"],
                              "checkins": res["venue"]["stats"]["checkinsCount"], "lat": res["venue"]["location"]["lat"],
                              "lng": res["venue"]["location"]["lng"]}, index=[v]))
        # Get the next venues
        next_venues = client.venues.nextvenues(v)
        for nv in next_venues['nextVenues']['items']:
            # Check the next venue is within bounding box
            if ((nv["location"]["lat"] > berlin_bbox[1]) & (nv["location"]["lat"] < berlin_bbox[3]) & 
                (nv["location"]["lng"] > berlin_bbox[0]) & (nv["location"]["lng"] < berlin_bbox[2])):
                # Add next venues to crawl list if not visited yet
                if (nv["id"] not in done) & (nv["id"] not in to_crawl) & (nv["id"] not in new_crawl):
                    new_crawl.append(nv["id"])
                links.append((v, nv["id"]))
        done.append(v)
    to_crawl = new_crawl
```

We can construct a directed graph from the data, using the locations as nodes and the adding the links between locations.

```python
labels = venues["name"].to_dict()
```

```python
import networkx as nx
G = nx.DiGraph()
G.add_nodes_from(venues.index)
for f,t in links:
    G.add_edge(f, t)
```

```python
print nx.info(G)
```

We can compute the **betweenness centrality** and the **pagerank** of the nodes. Betweenness centrality refers "is an indicator of a node's centrality in a network. It is equal to the number of shortest paths from all vertices to all others that pass through that node" (http://en.wikipedia.org/wiki/Betweenness_centrality). Pagerank (originally designed as an algorithm to rank web pages), "computes a ranking of the nodes in the graph based on the structure of the incoming links".  

```python
pagerank = nx.pagerank(G)
betweenness = nx.betweenness_centrality(G)
venues['pagerank'] = [pagerank[n] for n in venues.index]
venues['betweenness'] = [betweenness[n] for n in venues.index]
```

# Let's plot the top locations according to our different metrics

```python
%pylab inline
import matplotlib.pyplot as plt
def plot_top_locations(venues, column, n=20):
    fig = plt.figure(figsize=(8, 6), dpi=150)
    ax = fig.add_subplot(111)
    venues.sort(column, inplace=True)
    venues.set_index('name')[-20:][column].plot(kind='barh', color='r', alpha=0.5)
    ax.set_ylabel('location', fontsize=15)
    ax.set_xlabel(column, fontsize=15)
    ax.set_title('Top 20 locations by {}'.format(column), fontsize=15);
```

```python
plot_top_locations(venues, 'users')
plot_top_locations(venues, 'checkins')
plot_top_locations(venues, 'betweenness')
plot_top_locations(venues, 'pagerank')
```
