```python
import foursquare
import pandas as pd

CLIENT_ID = ""
CLIENT_SECRET = ""
client = foursquare.Foursquare(client_id=CLIENT_ID, client_secret=CLIENT_SECRET)
```

```python
ne = '52.55,13.47'
sw = '52.46,13.37'
#categories = '4bf58dd8d48988d116941735,4bf58dd8d48988d11f941735'
categories = '4d4b7105d754a06376d81259,4bf58dd8d48988d116941735,52e81612bcbc57f1066b7a0d,' \
             '4bf58dd8d48988d117941735,4bf58dd8d48988d11e941735,4bf58dd8d48988d118941735,' \
             '4bf58dd8d48988d120941735,4bf58dd8d48988d121941735,4bf58dd8d48988d11f941735,' \
             '4bf58dd8d48988d11b941735,4bf58dd8d48988d11d941735,4bf58dd8d48988d122941735,4bf58dd8d48988d123941735'
```

```python
# Look for categories 'bar' and 'nightclub'
res = client.venues.search(params={'ne': ne, 'sw': sw, 'limit': 50, 'intent': 'browse',
                                   'categoryId': categories})
```

```python
venues = pd.DataFrame()
for v in res['venues']:
    venues = venues.append(pd.DataFrame({"name": v["name"], "users": v["stats"]["usersCount"],
                              "checkins": v["stats"]["checkinsCount"], "lat": v["location"]["lat"],
                              "lng": v["location"]["lng"]}, index=[v['id']]))
```

Let's crawl these locations for the next venues to increase our database.

```python
to_crawl = venues.index
new_crawl = []
links = []
done = []
depth = 10
```

```python
for i in xrange(depth):
    new_crawl = []
    print "Step {} / {}: {} locations and {} links. {} venues to go.".format(
        i+1, depth, len(venues), len(links), len(to_crawl))
    for vi in to_crawl:
        next_venues = client.venues.nextvenues(vi)
        for nv in next_venues['nextVenues']['items']:
            # Check if it is bar or nightclub
            cats = [cat['id'] for cat in nv['categories']]
            if set(cats).intersection(set(categories.split(','))):
                # Check the next venue is within bounding box
                if ((nv["location"]["lat"] > float(sw.split(',')[0])) & (nv["location"]["lat"] < float(ne.split(',')[0])) & 
                    (nv["location"]["lng"] > float(sw.split(',')[1])) & (nv["location"]["lng"] < float(ne.split(',')[1]))):
                    if nv['id'] not in venues:
                        venues = venues.append(
                            pd.DataFrame({"name": nv["name"], "users": nv["stats"]["usersCount"],
                                          "checkins": nv["stats"]["checkinsCount"], "lat": nv["location"]["lat"],
                                          "lng": nv["location"]["lng"]},
                                          'likes': client.venues.likes(nv['id'])['likes']['count'], index=[nv['id']]))
                    if (nv["id"] not in done) & (nv["id"] not in to_crawl) & (nv["id"] not in new_crawl):
                        new_crawl.append(nv["id"])
                    links.append((vi, nv["id"]))
        done.append(vi)
    to_crawl = new_crawl
```

```python
venues = venues.reset_index().drop_duplicates(cols='index',take_last=True).set_index('index')
venues.describe()
```

```python
import networkx as nx
G = nx.DiGraph()
G.add_nodes_from(venues.index)
for f,t in links:
    G.add_edge(f, t)
print nx.info(G)
```

```python
pagerank = nx.pagerank(G)
betweenness = nx.betweenness_centrality(G)
centrality = nx.degree_centrality(G)
venues['pagerank'] = pd.Series(pagerank, index=venues.index)
venues['betweenness'] = pd.Series(betweenness, index=venues.index)
venues['centrality'] = pd.Series(centrality, index=venues.index)
```

```python
%pylab inline
import matplotlib.pyplot as plt
def plot_top_locations(venues, column, n=20):
    fig = plt.figure(figsize=(8, 6), dpi=150)
    ax = fig.add_subplot(111)
    venues.sort(column, inplace=True)
    venues.set_index('name')[-n:][column].plot(kind='barh', color='r', alpha=0.5)
    ax.set_ylabel('location', fontsize=15)
    ax.set_xlabel(column, fontsize=15)
    ax.set_title('Top 20 locations by {}'.format(column), fontsize=15);
```

```python
plot_top_locations(venues, 'users', 20)
plot_top_locations(venues, 'checkins', 20)
plot_top_locations(venues, 'betweenness', 20)
plot_top_locations(venues, 'centrality', 20)
plot_top_locations(venues, 'pagerank', 20)
```

```python
labels = venues["name"].to_dict()
fig = plt.figure(figsize=(15, 15), dpi=150)
graph_pos = nx.spring_layout(G, k=0.08, iterations=1000)
nodesize = dict()
for k, v in venues.pagerank.iteritems():
    nodesize[k] = v * 100000
for n in G.nodes():
    nx.draw_networkx_nodes(G, graph_pos, nodelist=[n], node_size=nodesize[n], alpha=0.5, node_color='green')
nx.draw_networkx_edges(G, graph_pos, width=0.8, alpha=0.2, edge_color='blue')
nx.draw_networkx_labels(G, graph_pos, labels=labels, font_size=8)
plt.axis('off');
```
