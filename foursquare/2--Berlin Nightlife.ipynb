# Berlin's Nightlife: The Hotspots According to Foursquare

In this notebook we are going to crawl Foursquare data for nightlife venues in Berlin and explore this data.

First, we need access to Foursquare's API.

```python
from cPickle import load
import foursquare
import pandas as pd

keys = load(file('foursquare_keys.pkl'))
CLIENT_ID = client_id=keys['client_id']
CLIENT_SECRET = keys['client_secret']
client = foursquare.Foursquare(client_id=CLIENT_ID, client_secret=CLIENT_SECRET)
```

We then define a location bounding box that roughly covers Mitte, Prenzlauer Berg, Friedrichshain, Kreuzberg, and Neukoelln. `ne` and `sw` are the north-east and south-west longitude/latitude coordinates of the bounding box, respectively.

```python
ne = '52.55,13.47'
sw = '52.46,13.37'
```

Now we define which [Foursquare venue categories](https://developer.foursquare.com/categorytree) we want to retrieve data from. Here I choose: Nightlife Spot, Bar, Beach Bar, Beer Garden, Cocktail Bar, Dive Bar, Karaoke Bar, Lounge, Nightclub, Pub, Sports Bar, Whisky Bar, Wine Bar.

```python
categories = '4d4b7105d754a06376d81259,4bf58dd8d48988d116941735,52e81612bcbc57f1066b7a0d,' \
             '4bf58dd8d48988d117941735,4bf58dd8d48988d11e941735,4bf58dd8d48988d118941735,' \
             '4bf58dd8d48988d120941735,4bf58dd8d48988d121941735,4bf58dd8d48988d11f941735,' \
             '4bf58dd8d48988d11b941735,4bf58dd8d48988d11d941735,4bf58dd8d48988d122941735,4bf58dd8d48988d123941735'
```

Let's query Foursquare; We will get back a total of 50 nighlife spots within the bounding box.

```python
res = client.venues.search(params={'ne': ne, 'sw': sw, 'limit': 50, 'intent': 'browse', 'categoryId': categories})
```

We now can store the entries we are interested in, in a pandas dataframe. For each venue, I store *name*, *# users*, *# checkins*, *latitude*, *longitude*, *# likes*.

```python
venues = pd.DataFrame()
for v in res['venues']:
    venues = venues.append(
        pd.DataFrame({"name": v["name"], "users": v["stats"]["usersCount"],
                      "checkins": v["stats"]["checkinsCount"], "lat": v["location"]["lat"],
                      "lng": v["location"]["lng"], 'likes': client.venues.likes(v['id'])['likes']['count']},
                      index=[v['id']]))
```

We now can use these 50 locaitons and for each one retrieve the next venues (https://developer.foursquare.com/docs/venues/nextvenues) to increase our database. We do this recursively, until a certain depth has been reached or there are no more locations to go to.

```python
to_crawl = venues.index
new_crawl = []
links = []
done = []
depth = 10
for i in xrange(depth):
    new_crawl = []
    print "Step {} / {}: {} locations and {} links. {} venues to go.".format(
        i+1, depth, len(venues), len(links), len(to_crawl))
    for vi in to_crawl:
        next_venues = client.venues.nextvenues(vi)
        for nv in next_venues['nextVenues']['items']:
            # Check if it is bar or nightclub
            cats = [cat['id'] for cat in nv['categories']]
            if set(cats).intersection(set(categories.split(','))):
                # Check the next venue is within bounding box
                if ((nv["location"]["lat"] > float(sw.split(',')[0])) & (nv["location"]["lat"] < float(ne.split(',')[0])) & 
                    (nv["location"]["lng"] > float(sw.split(',')[1])) & (nv["location"]["lng"] < float(ne.split(',')[1]))):
                    if nv['id'] not in venues:
                        venues = venues.append(
                            pd.DataFrame({"name": nv["name"], "users": nv["stats"]["usersCount"],
                                          "checkins": nv["stats"]["checkinsCount"], "lat": nv["location"]["lat"],
                                          "lng": nv["location"]["lng"],
                                          'likes': client.venues.likes(nv['id'])['likes']['count']}, index=[nv['id']]))
                    if (nv["id"] not in done) & (nv["id"] not in to_crawl) & (nv["id"] not in new_crawl):
                        new_crawl.append(nv["id"])
                    links.append((vi, nv["id"]))
        done.append(vi)
    to_crawl = new_crawl
```

Nice! Now I am adding a *likes_ratio* column into the dataframe (basically normalizes number of likes by number of users), and I remove potential duplicates in the dataframe.

```python
venues['likes_ratio'] = pd.Series(venues['likes'].values / venues['users'].values.astype('float'), index=venues.index)
venues = venues.reset_index().drop_duplicates(cols='index',take_last=True).set_index('index')
```

Awesome, moving on. We are going to use Python's `netwworkx` library in order to construct a directed graph from our data.

```python
import networkx as nx
G = nx.DiGraph()
G.add_nodes_from(venues.index)
for f,t in links:
    G.add_edge(f, t)
print nx.info(G)
```

Using this graph we are able to compute a few network metrics. Here I compute **degree centrality**, **betweenness centrality** and **pagerank**.

```python
pagerank = nx.pagerank(G)
betweenness = nx.betweenness_centrality(G)
centrality = nx.degree_centrality(G)
venues['pagerank'] = pd.Series(pagerank, index=venues.index)
venues['betweenness'] = pd.Series(betweenness, index=venues.index)
venues['centrality'] = pd.Series(centrality, index=venues.index)
```

# Let's plot the Top Locations according to our metrics!

```python
%pylab inline
import matplotlib.pyplot as plt
def plot_top_locations(venues, column, n=20, ax=None):
    venues.sort(column, inplace=True)
    venues.set_index('name')[-n:][column].plot(kind='barh', color='r', alpha=0.5, fontsize=15)
    ax.set_ylabel('')
    ax.set_title('Top {} locations by {}'.format(n, column), fontsize=20);
```

```python
columns_to_plot = ['users', 'checkins', 'betweenness', 'centrality', 'pagerank', 'likes', 'likes_ratio']
fig = plt.figure(figsize=(10, 40), dpi=150)
for i in xrange(len(columns_to_plot)):
    ax = fig.add_subplot(7, 1, i+1)
    plot_top_locations(venues, columns_to_plot[i], 20, ax)
plt.tight_layout()
```

We can also visualize the whole network. The size of the nodes is with respect to the pagerank value.

```python
labels = venues["name"].to_dict()
fig = plt.figure(figsize=(15, 15), dpi=150)
graph_pos = nx.spring_layout(G, k=0.08, iterations=1000)
nodesize = dict()
for k, v in venues.pagerank.iteritems():
    nodesize[k] = v * 100000
for n in G.nodes():
    nx.draw_networkx_nodes(G, graph_pos, nodelist=[n], node_size=nodesize[n], alpha=0.5, node_color='green')
nx.draw_networkx_edges(G, graph_pos, width=0.8, alpha=0.2, edge_color='blue')
nx.draw_networkx_labels(G, graph_pos, labels=labels, font_size=8)
plt.axis('off');
```
