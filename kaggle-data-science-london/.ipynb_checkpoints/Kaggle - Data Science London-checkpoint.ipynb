Kaggle: Data Science London
===========================

In this notebook I am going to show how to build a simple model for the [**Kaggle Data Science London**](https://www.kaggle.com/c/data-science-london-scikit-learn) competition using [scikit-learn](http://scikit-learn.org/stable/).

The data set at hand is synthetic data that consists of 40 features, representing objects from two classes (labeled as 0 or 1). Hence, the present machine learning task is a binary classification problem. The training set has 1000 samples and the testing set has 9000. Because the data is synthetic, the features do not represent anything meaningful or interpretable.

First, let's import some libraries that we will be using along the way:

```python
import csv
import numpy as np
import pandas as pd
import pylab as pl

from sklearn.cross_validation import cross_val_score
from sklearn.decomposition import PCA
from sklearn.grid_search import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

Let's start by loading the data. It consists of the training samples (*'train.csv'*), the training targets (*'trainLabels.csv'*), and the test samples (*'test.csv'*). We can read the data using Pandas and convert it to Numpy arrays:

```python
X_train = pd.read_csv('train.csv', header=None).as_matrix()
y_train = np.squeeze(pd.read_csv('trainLabels.csv', header=None).as_matrix())
X_test = pd.read_csv('test.csv', header=None).as_matrix()
```

Principal Component Analysis (PCA)
----------------------------------

To get a better sense of how the data looks like, let's explore it. Therefor, I project the data to 2 principal components in order to visualize it. We can make use of the large test set size and incorporate it in the our PCA:

```python
pca = PCA(n_components=2, whiten=True)
X_pca = pca.fit_transform(np.r_[X_train, X_test])
```

```python
%matplotlib inline
fig, ax = pl.subplots(1, figsize=(8, 8))
c1 = X_pca[y_train == 0, :]
c2 = X_pca[y_train == 1, :]
ax.scatter(c1[:, 0], c1[:, 1], color='r', marker='*', s=20)
ax.scatter(c2[:, 0], c2[:, 1], color='g', marker='o', s=20)
ax.set_xlabel('princomp 1')
ax.set_ylabel('princomp 2')
ax.grid('on', 'both')
```

Doesn't look very separable using only the first two components, does it? Let's have a look at the explained ratio of the principal components:

```python
pca = PCA(whiten=True)
X_pca = pca.fit_transform(np.r_[X_train, X_test])
fig, ax = pl.subplots(1, figsize=(8, 5))
ax.plot(range(X_pca.shape[1]), pca.explained_variance_ratio_, 'k', linewidth=2.5)
ax.set_xlabel('# princomps')
ax.set_ylabel('explained variance ratio')
ax.grid('on', 'both')
```

We can also plot the cumulative sum of the principal components' explained variance ratio:

```python
fig, ax = pl.subplots(1, figsize=(8, 5))
ax.plot(range(X_pca.shape[1]), np.cumsum(pca.explained_variance_ratio_), 'k', linewidth=2.5)
ax.set_xlabel('# princomps')
ax.set_ylabel('cumsum explained variance ratio')
ax.grid('on', 'both')
```

Inspecting both graphs we can observe mainly two major insights that might be helpful for preprocessing of the data:

1. The last 3 components do not provide information. 100% of the variance is explained by using the first 37 principal components.
2. Most of the variance (> 80%) is explained by the first 12 pricipal components.

Using Support Vector Machines
-----------------------------

Ok, let's jump straight into training a classifier. There are many methods to try out here, including Logistic Regression, LDA, Random Forests, etc. Another straight-forward choice (which I make here) are Support Vector Machines (SVMs). In general, the first step I would take after deciding on a method, is to perform model selection by tuning the crucial hyperparameters of the model. In particular, SVMs have two hyperparameters that need to be tuned: **C** and **gamma**. You could also try out different kernels, but here I will stick simply to the default RBF kernel and do a gridsearch over C and gamma.
In addition, I will make use of the scikit-learn *Pipeline* object, because I intend to first scale the data to zero mean and unit variance in each cross-validation fold.

```python
pip = Pipeline([('scaler', StandardScaler()), ('clf', SVC(kernel='rbf'))])
C_range = np.logspace(-2, 3, 20)
gamma_range = np.concatenate((np.asarray([0]), np.logspace(-8, 0, 20)))
params = {'clf__C': C_range,
          'clf__gamma': gamma_range}
gs = GridSearchCV(pip, params, cv=5, n_jobs=-1)
gs.fit(X_train, y_train);
```

Now that the gridsearch over the specified parameter range is completed, we can inspect the best validation score and the best estimator:

```python
gs.best_score_, gs.best_estimator_
```

To get a better overview over the performace of the different models, I define a helper function that plots the validation score for each pair of hyperparameter values:

```python
def plot_grid(grid_search):
    """ Plot the validation score for each value of C and gamma from a gridsearch on an SVM.
        
        :param grid_search: A scikit-learn GridSearchCV object.
    
    """
    grid_scores = grid_search.grid_scores_
    gammas, Cs, scores = [], [], []
    gamma_str, c_str = ('clf__gamma', 'clf__C') if isinstance(gs.best_estimator_, Pipeline) else ('gamma', 'C')
    for grid in grid_scores:
        gammas.append(grid.parameters[gamma_str])
        Cs.append(grid.parameters[c_str])
        scores.append(grid.mean_validation_score)
    gammas = np.asarray(gammas)
    Cs = np.asarray(Cs)
    scores = np.asarray(scores)
    scores = scores.reshape(len(C_range), len(gamma_range))
    pl.figure(figsize=(10,10))
    pl.imshow(scores, interpolation='nearest', cmap=pl.cm.spectral)
    pl.xlabel('gamma')
    pl.ylabel('C')
    pl.xticks(np.arange(len(gamma_range)), gamma_range, rotation=45)
    pl.yticks(np.arange(len(C_range)), C_range)
    pl.colorbar()
```

```python
plot_grid(gs)
```

The best score resulting from this first grid search is around 87%. The weird thing is (and it took me a while to figure this out) that using an SVM with default hyperparameters already gives around 91% accuracy:

```python
np.mean(cross_val_score(SVC(), X_train, y_train, cv=5))
```

Somehow scaling the data to zero mean and unit variance decreases the quality of the model. This is probably because there is noise in the data, leading to a worse performance after scaling. Hence, let's perform the same grid search again, this time without standard scaling:

```python
clf = SVC(kernel='rbf')
C_range = np.logspace(-2, 3, 20)
gamma_range = np.concatenate((np.asarray([0]), np.logspace(-8, 0, 20)))
params = {'C': C_range,
          'gamma': gamma_range
}
gs = GridSearchCV(clf, params, cv=5, n_jobs=-1)
gs.fit(X_train, y_train);
```

```python
plot_grid(gs)
```

```python
gs.best_score_, gs.best_estimator_
```

Indeed, without prior scaling of the data, our model now performs better. Based on the grid search plot, we can run a more detailed search around the values that resulted in the best validation scores. There seem to be two potentially good parameter ranges, one where gamma is 0, and one where gamma is somewhere arounf 0.002.

Let's explore the first possibility one in more detail by fixing gamma to 0 and varying C:

```python
clf = SVC(kernel='rbf')
C_range = np.linspace(0.8, 5, 30)
params = {'C': C_range}
gs = GridSearchCV(clf, params, cv=5, n_jobs=-1)
gs.fit(X_train, y_train)
Cs, scores = [], []
for grid in gs.grid_scores_:
    Cs.append(grid.parameters['C'])
    scores.append(grid.mean_validation_score)        
Cs = np.asarray(Cs)
scores = np.asarray(scores)
pl.figure(figsize=(8,4))
pl.plot(Cs, scores, 'k', linewidth=2.5)
pl.xlabel('C')
pl.ylabel('Score')
pl.xticks(C_range, rotation=45)
pl.grid('on', 'both')
```

```python
gs.best_score_, gs.best_estimator_
```

This classifier has a C=.94 and gamma=0.0 with a validation score of 91.6% accuracy.

Let's explore the second potentially good range in our gridsearch space:

```python
clf = SVC(kernel='rbf')
C_range = np.logspace(0, 3, 20)
gamma_range =  np.linspace(0.01, 0.02, 20)
params = {'C': C_range,
          'gamma': gamma_range
}
gs = GridSearchCV(clf, params, cv=5, n_jobs=-1)
gs.fit(X_train, y_train);
```

```python
plot_grid(gs)
```

```python
gs.best_score_, gs.best_estimator_
```

This classifier (gamma=0.015, C=2.1) reaches 92.4% accuray. Slightly better, but not much. An observation: The value of C seems to be invariant to the performance, as long as it is big enough. Let's test that this is indeed the case by arbitrarily setting its value to 10000:

```python
clf = SVC(kernel='rbf', gamma=0.015, C=10000)
cross_val_score(clf, X_train, y_train, cv=5, n_jobs=-1).mean()
```

Using Principal Components
--------------------------

Let's test whether can improve our model by using principal components as features. Therefor, we use the PCA as a first step in a scikit-learn Pipeline object, and perform a gridsearch over all the possible numbers of principal components, as well as over reasonable ranges of gamma and C:

```python
pca = PCA(whiten=True)
pip = Pipeline([('pca', pca), ('clf', SVC(kernel='rbf'))])
params = {'pca__n_components': np.arange(2, 40),
          'clf__gamma': np.linspace(0, 0.02, 10),
          'clf__C': np.linspace(0.9, 3, 20)}
gs = GridSearchCV(pip, params, cv=5, n_jobs=-1)
gs.fit(X_train, y_train);
```

```python
gs.best_estimator_, gs.best_score_
```

We were able to boost our model to 94.7% by using the first 12 principal components as features!

Predicting
----------

Now that we have a decent classifier, we can predict the labels  of the test samples, and save these in a CSV-file that we can upload to the competitions' page on Kaggle to see how we perform on the leaderboard (spoiler: we reach 0.94188).

```python
def save_predictions(preds):
    ids = np.arange(1, 9001)
    with open('predictions.csv', 'w') as fp:
        a = csv.writer(fp, delimiter=',')
        header = ['Id', 'Solution'],
        a.writerows(header)
        a.writerows(zip(ids, preds))
```

```python
save_predictions(gs.predict(X_test))
```

Using Gaussian Mixture Model
----------------------------

The following part is not my own idea, but merely something that has been discussed in the competitions' forum, and thus I wanted to try it out here. If you plot the distributions of the principal components, it seems like they all are (except the las three components) Gaussian distributed. Therefore, it has been proposed to use a Gaussian Mixture Model (GMM) on the principal components and feed the predicted probabilities into a Support Vector Machine as features for classification.
I am not going to perform a detailed gridsearch over the parameters of the GMM, but simply use what has been reported to work well (n_components = 4, covariance_type = 'full').

```python
from sklearn.mixture import GMM
```

```python
pca = PCA(n_components=37, whiten=True)
X_pca = pca.fit_transform(X_train)
gmm = GMM(n_components=4, covariance_type='full')
gmm.fit(X_pca)
probs = gmm.predict_proba(X_pca)
```

```python
cross_val_score(SVC(C=3.), probs, y_train, cv=5, n_jobs=-1).mean()
```

Indeed, using the GMM we reach a validation score as high as 99%! Hence, we can train our SVM on the probabilities of the GMM, and predict the test set samples using the same steps as during training:

1. PCA
2. GMM
3. SVM

```python
clf.fit(probs, y_train)
X_pca = pca.transform(X_test)
probs = gmm.predict_proba(X_pca)
preds = clf.predict(probs)
save_predictions(clf.predict(probs))
```
