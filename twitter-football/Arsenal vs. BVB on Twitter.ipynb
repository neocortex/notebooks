```python
texts = [t.encode('ascii', 'ignore') for t in df.text]
text = ' '.join(texts)
```

```python
import nltk
nltk.data.path = ['/home/rafael/libraries/nltk/nltk_data']
words = nltk.word_tokenize(text)
```

### Only ASCII encoded words

```python
ascii_words = []
for word in words:
    try:
        w = word.encode('ascii', 'strict')
    except UnicodeEncodeError:
        continue
    else:
        ascii_words.append(w)
```

### Remove punctuation

```python
import string
words = [word.translate(None, string.punctuation) for word in ascii_words]
```

### Remove single characters

```python
words = [word for word in words if len(word) > 1]
```

```python
import sys
sys.path.append('/home/rafael/libraries/word_cloud/')
import wordcloud

words = wordcloud.process_text(' '.join(words), max_features=2000)
elems = wordcloud.fit_words(words, width=1200, height=800, margin=5, ranks_only=False, prefer_horiz=0.90)
```

```python
wordcloud.draw(elems, 'wordcloud.png', width=1200, height=800)
```
